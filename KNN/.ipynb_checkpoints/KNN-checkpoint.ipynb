```python
new_data = my_data.drop('Pregnancies', axis=1)

#my_data adlı DataFrame'den Pregnancies adlı sütunu kaldırır 
#ve bu işlem sonucunda oluşan yeni DataFrame'i new_data adlı değişkene atar
#axis=1: Bu parametre sütunun kaldırılacağını belirtir.
#axis=0 kullanılsaydı satırların kaldırılması anlamına gelir
```

```python
new_data1 = new_data.drop('SkinThickness', axis=1)
new_data2 = new_data1.drop('BloodPressure', axis=1)
new_data3 = new_data2.drop('BMI', axis=1)
new_data4 = new_data3.drop('Age', axis=1)
new_data5 = new_data4.drop('DiabetesPedigreeFunction', axis=1)
```

```python
new_data5
```

```python
my_newdata = new_data5.drop('Outcome',axis=1)
```

```python
my_newdata
```

```python
user_data = [160,35]

#user_data = [160, 35] ifadesi, user_data adlı bir liste oluşturur ve bu listeye iki değer atar: 160 ve 35.
#u tür bir liste, kullanıcı özelliklerini, belirli bir veri noktasını veya analiz ve görselleştirme 
#işlemlerinde kullanılabilecek herhangi bir veri kümesini temsil etmek için kullanılabilir.
```

```python
columns = list(my_newdata)

#sütun adlarına kolayca erişmek ve bunları kullanarak çeşitli işlemler yapmak için kullanılır
#Bu ifade, my_newdata adlı DataFrame'in sütun adlarını içeren bir liste oluşturur
```

```python
columns
```

```python
data_arr = []
dist_arr = []
```

```python
for i in range(0,20):
    for j in columns:
        print(my_newdata[j][i])
        elem = my_newdata[j][i]
        data_arr.append(elem)
    print(data_arr)    
    dist = ((user_data[0]-data_arr[0]) **2 + (user_data[1]-data_arr[1]) **2) **0.5
    dist_arr.append(dist)
    data_arr = []
```

```python
dist_arr
```

```python
result = pd.DataFrame(dist_arr)
```
