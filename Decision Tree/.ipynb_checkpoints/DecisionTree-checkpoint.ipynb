```python
import pandas as pd
```

```python
df = pd.read_csv("salaries.csv")
```

```python
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']

#df isimli bir veri çerçevesinden 'salary_more_then_100k' isimli sütunu çıkartır.
#axis='columns', işlemin sütunlar üzerinde yapılacağını belirtir.
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder

#LabelEncoder sınıfını içe aktarır
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()

#LabelEncoder sınıfını kullanarak üç 
#farklı kategorik sütunu (company, job, degree) etiketlemek için hazırlanmıştır
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()

#fit_transform metodu, önce fit (veriyi analiz edip uygun etiketleri belirleme) sonra
#transform (bu etiketleri kullanarak veriyi dönüştürme) işlemlerini gerçekleştirir.
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis = 'columns')
inputs_n

#drop kaldır,çıkart
#inputs adlı veri çerçevesinden company, job ve degree sütunlarını çıkartarak geriye 
#kalan sütunları inputs_n adlı yeni bir veri çerçevesine atar
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)

#inputs_n veri çerçevesindeki özellikleri (bağımsız değişkenler) kullanarak target (bağımlı değişken)
#üzerinde model (bir DecisionTreeClassifier nesnesi) ile eğitme işlemini gerçekleştirir
```

```python
model.score(inputs_n,target)

#Bu kod modelin tüm veri kümesi üzerindeki doğruluk skorunu hesaplar
#score metodu, modelin doğruluğunu hesaplamak için kullanılır
```

```python
new_data = [[2,0,1]]
prediction = model.predict(new_data)

#[2, 2, 1] özellik vektörü şu anlama gelir:

#company_n = 2 (google)
#job_n = 2 (sale executive).
#degree_n = 1 (masters).
```

```python
print(f'Tahmin: {prediction[0]}')
```
